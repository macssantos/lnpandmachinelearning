```python
import random

num_aleatorio = random.randint(0,100)

def numaleatorio(num):
  if num == num_aleatorio:
    return 0
  elif num > num_aleatorio:
    return 1
  else:
    return -1


num = int(input('Adivinhe qual o número'))
result = numaleatorio(num)

while result != 0:
  if result == 1:
    print('O número secreto é MENOR que seu palpite. Tente novamente!')
  else:
    print('O número secreto é MAIOR que seu palpite. Tente novamente!')


  num = int(input('Adivinhe qual o número'))
  result = numaleatorio(num)

print('Parabéns! Você acertou o número!')
```
